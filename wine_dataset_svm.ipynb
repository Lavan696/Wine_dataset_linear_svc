```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_wine
```

```python
wine_dataset = load_wine(as_frame = True)
```

```python
wine_dataset
```

```python
X = wine_dataset.data
X
```

```python
y = wine_dataset.target
y
```

```python
X.info()
```

```python
X.describe()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 44)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
X_train
```

```python
len(y_train)
```

```python
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
poly_lin_svm_clf = make_pipeline(
    PolynomialFeatures(degree = 2),
    StandardScaler(),
    LinearSVC(C = 10, max_iter = 10_000, random_state = 42,dual = False)
)
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=44)
scores = cross_val_score(poly_lin_svm_clf, X_train, y_train, cv=skf)
scores
```

```python
scores.mean()
```

```python
score = cross_val_score(poly_lin_svm_clf, X_train, y_train, cv=10)
print(f"Mean accuracy: {scores.mean():.3f} ± {scores.std():.3f}")
```

```python
poly_lin_svm_clf.fit(X_train,y_train)
```

```python
y_predicted = poly_lin_svm_clf.predict(X_test)
y_predicted
```

```python
poly_lin_svm_clf.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predicted)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=poly_lin_svm_clf.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```
